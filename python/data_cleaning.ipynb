## Data Cleaning in Python

## Import

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/customer_shopping_behavior.csv")
```

```python
df
```

## Data Cleaning

```python
df.info()
```

```python
string_cols = df.select_dtypes(include = ['str']).columns
```

```python
df[string_cols] = df[string_cols].apply(lambda x: x.str.strip())
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median() ))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.info()
```

```python
df['age'].describe()
```

* 18-27 = Young Adult
* 28-39 = Adult
* 40-59 = Middle Age
* 60+ = Senior

```python
bins = [18, 28, 40, 60, np.inf]
labels = ["Young Adult", "Adult", "Middle Age", "Senior"]
df["age_group"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels,
    right=False,          # [18,28), [28,40), [40,60), [60,inf)
    include_lowest=True
)
```

```python
df
```

```python
# create new column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[df['promo_code_used']!= df['discount_applied']]
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
df.to_csv('./da_ready_data.csv', index=False)
```
